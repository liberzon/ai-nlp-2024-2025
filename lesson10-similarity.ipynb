```python
import numpy as np
from numpy.linalg import norm
```

```python
def cosine_distance(A, B):
    return 1.0-np.dot(A,B)/(norm(A)*norm(B))
```

```python
cosine_distance(v_king, v_queen)
```

```python
model.most_similar("queen")  # show words that similar to word
```

```python
cosine_distance(v_king-v_man+v_woman, v_queen)
```
